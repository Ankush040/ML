# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Importing dataset given by the TechWorks

```python
df=pd.read_csv("C:/Users/ankus/Downloads/ML case Study (1).csv",header=0)
college=pd.read_csv("C:/Users/ankus/Downloads/Colleges (1).csv",header=0)
city=pd.read_csv("C:/Users/ankus/Downloads/cities (2).csv",header=0)
```

### Seeing sample of dataset

```python
df.head()
```

```python
Tier1=college["Tier 1"].tolist()
Tier2=college["Tier 2"].tolist()
Tier3=college["Tier 3"].tolist()
Metro_city=city["Metrio City"].tolist()
Non_metro_city=city["non-metro cities"].tolist()
```

#### Converting College column from categorical to numerical data type

```python
for item in df.College:
    if item in Tier1:
        df["College"].replace(item,1,inplace=True)
    elif item in Tier2:
        df["College"].replace(item,2,inplace=True)
    elif item in Tier3:
        df["College"].replace(item,3,inplace=True)
```

```python
Metro_city,Non_metro_city
```

```python
df.head()
```

#### Converting City column data to numerical type

```python
for i in df.City:
    if i in Metro_city:
        df["City"].replace(i,1,inplace=True)
    elif i in Non_metro_city:
        df["City"].replace(i,0,inplace=True)
```

```python
df.head()
```

#### Creating dummies for categorical columns 

```python
df=pd.get_dummies(df)
```

#### Deleting "Role_Manager" 

```python
del df["Role_Manager"]
```

## Creating indenpendent and dependent varibles 

```python
x=df.loc[:,df.columns!="CTC"]
```

```python
y=df["CTC"]
```

## Importing statsmodels to train our model 

```python
import statsmodels.api as sn
```

Adding constant to indenpendent varible as statsmodel does not include it

```python
X=sn.add_constant(x)
```

### Creating object salary and fitting it using statsmodel.api

```python
salary=sn.OLS(y,X).fit()
```

#### Seeing the summary of our model given by statsmodels

```python
salary.summary()
```

# Importing Linear Regression from sklearn

```python
from sklearn.linear_model import LinearRegression
```

## Spliting data for test and train purpose 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

#### Creating salary_LR object

```python
salary_LR = LinearRegression()
```

## Fitting our model by using train varibles

```python
salary_LR.fit(X_train,y_train)
```

```python
print(salary_LR.intercept_)
```

#### Predicting dependent varible using test data

```python
y_pred=salary_LR.predict(X_test)
```

#### Predicting dependent varible using train data

```python
y_train_pred=salary_LR.predict(X_train)
```

#### Importing sklearn.metrics for r2_score for checking our model performance

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
r2_score(y_train,y_train_pred)
```

# Trying Ridge Regression 

#### Importing preprocessing

```python
from sklearn import preprocessing
```

#### Creating object scaler to standardzied varibles

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

#### Standardzing varibles using scaler object 

```python
X_train_s=scaler.transform(X_train)
```

```python
X_test_s=scaler.transform(X_test)
```

### Importing Ridge

```python
from sklearn.linear_model import Ridge
```

### Creating object lm_r and fitting our model 

```python
lm_r=Ridge(alpha = 0.8)
```

```python
lm_r.fit(X_train_s,y_train)
```

### Checking model performance on test data

```python
r2_score(y_test,lm_r.predict(X_test_s))
```

### Importing validation curve and trying our model for wide range of data to increase accuracy and repeating processes

```python
from sklearn.model_selection import validation_curve
```

```python
param_range=np.logspace(-2,8,100)
```

```python
train_scores, test_scores= validation_curve(Ridge(),X_train_s,y_train,param_name="alpha",param_range=param_range,scoring="r2")
```

```python
print(train_scores)
print(test_scores)
```

```python
train_mean=np.mean(train_scores,axis=1)
test_mean=np.mean(test_scores,axis=1)
```

```python
max(train_mean)
```

```python
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range),y=test_mean)
```

### Finding the location where our model have maximum efficiency

```python
np.where(test_mean==max(test_mean))
```

```python
param_range[26]
```

#### Again fitting our model and finding r2_score

```python
lm_r_best=Ridge(alpha=param_range[26])
```

```python
lm_r_best.fit(X_train_s,y_train)
```

```python
r2_score(y_test,lm_r_best.predict(X_test_s))
```
